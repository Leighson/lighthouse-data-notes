# Matrix Decomposition

### Decomposition of Images Using SVD

One of the applications of SVD is image compression. The key here is that a black and whtie image is simply a matrix where the numbers represent the intensity of a given pixel. The matrix can be decomposed just like any other.

```python
import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
```

```python
img = Image.open("./flower_square.jpeg")
imggray = img.convert('LA') 
plt.figure(figsize=(9, 6))
plt.imshow(imggray)
```

```python
imgmat = np.array(list(imggray.getdata(band=0)), float)
print(type(imgmat))
print(imgmat.shape)
imgmat.shape = (imggray.size[1], imggray.size[0])
print(imgmat.shape)

imgmat = np.matrix(imgmat)
print(type(imgmat))
print(imgmat.shape)

plt.figure(figsize=(9,6))
plt.imshow(imgmat, cmap='gray')
```

```python
U, sigma, V = np.linalg.svd(imgmat)
```

```python
reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
plt.imshow(reconstimg, cmap='gray')
```

```python
# we have replaced 1 from previous code by i
for i in range(2, 4):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```

```python
for i in range(5, 51, 5):
    reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
    plt.imshow(reconstimg, cmap='gray')
    title = "n = %s" % i
    plt.title(title)
    plt.show()
```
