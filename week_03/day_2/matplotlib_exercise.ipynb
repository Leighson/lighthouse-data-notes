# Sample Application of Visualization

## Stateful Approach

```python
import matplotlib.pyplot as plt
import pandas as pd
```

#### Imagine we were asked to find the richest country in the world on a per-person basis in thee sample dataset below:

```python
my_data = pd.read_csv('nations.csv')
```

```python
my_data['gdp_percap'].fillna(my_data['gdp_percap'].median(), inplace=True)
```

```python
my_data.groupby(['country']).mean()
```

```python
avg_gdp_percap = my_data.groupby(['country']).mean()['gdp_percap']
```

```python
top_five_countries = avg_gdp_percap.sort_values(ascending=False).head()
```

```python
china = my_data[my_data['country'] == 'Macao SAR, China']
```

```python
china['population'].iloc[0]
```

```python
plt.plot(china['year'], china['gdp_percap'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP Per Capita of Macao SAR, China')
```

```python
china.sort_values(by='year').plot.bar(x='year', y='gdp_percap')
```

```python
plt.subplot(311)
plt.title('GDP Per Capita')
plt.bar(china['year'], china['gdp_percap'])

plt.subplot(312)
plt.title('GDP in Billions')
plt.bar(china['year'], (china['population']*china['gdp_percap']/10**9))

plt.subplot(313)
plt.title('Population in Millions')
plt.bar(china['year'], china['population']/10**6)

plt.tight_layout()
```

```python
plt.bar(china['year'], (china['population']/china['population'].iloc[0]*400), color='r')

china_gdp = china['population'] * china['gdp_percap']
plt.bar(china['year'], china_gdp/(china_gdp.iloc[0]/400), color='g')

plt.bar(china['year'], china['gdp_percap'].iloc[0]/400, color='b')
    
plt.title('GDFP and Population Growth in China(first = 100)')
plt.legend(['Population', 'GDP', 'GDP Per Capita'], loc=4)
```
