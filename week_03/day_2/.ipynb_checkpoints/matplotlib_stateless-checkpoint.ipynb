```python
import matplotlib.pyplot as plt
import numpy as np 
from numpy import genfromtxt
%matplotlib inline
```

```python
# load data using its file name
filename = './GLB.Ts+dSST.csv'
data = genfromtxt(filename, delimiter=',')

# assign data to variables
year = data[1:, 0]
jan = data[1:, 1]
aug = data[1:, 8]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,5))

ax.plot(year, jan,
        label='January',
        c='steelblue',
        ls='-', lw=3,
        marker='.',
        ms=7,
        mfc='#50514F',
        zorder=2)

ax.plot(year, aug,
        label='August',
        c='#F25F5C',
        ls='--',
        lw=3,
        zorder=2)

# define legend formatting
# locations:
# - 1 = upper right
# - 2 = upper left
# - 3 = lower left
# - 4 = lower right
leg = ax.legend(loc=4, ncol=2, numpoints=1, fontsize=15, shadow=False)

# draw centerline at 0 temp
ax.axhline(0, lw=2, c='dimgray', ls='-', alpha=0.5, zorder=1) # plotting functions will be executed according the zorder

# other customizations
ax.set_ylabel('Temperature\nAnomaly [$^o$C]', fontsize=20)
ax.set_xlabel('Year', fontsize=20)

ax.set_title('Yearly Temperature Anomaly Means for January and August', fontsize=24, y=1.05)

# plot extends to year 2025 for extra white space on the right side
xticklabels = np.arange(2000, 2024, 4)
ax.xaxis.set_ticks(xticklabels)
ax.set_xticklabels(xticklabels, fontsize=15, rotation=45)

# naxis.set_ticks is inherited through set_nticklabels
# allows for specifity that set_nticklabels does not provide
yticklabels = np.arange(-1, 1.4, .5)
ax.yaxis.set_ticks(yticklabels)
ax.set_yticklabels(yticklabels, fontsize=15)

# though the range above dictates the possible intervals,
# set_nlim dictates what is seen on the graph
ax.set_xlim([2002, 2025])
ax.set_ylim([-1, 1])

# display grid
ax.grid()

leg.set_title(title='Month', prop=dict(size = 15))

# sets color around the graph
fig.patch.set_facecolor('lightsteelblue')

# saves figure to .png
fig.savefig('./jan_aug_tempanom.png', dpi=150, bbox_inches='tight', format='png', facecolor=fig.get_facecolor())
```
