```python
# import pandas
import pandas as pd  

# from apyori import apriori
from apyori import apriori
```

```python
# load data 
movie_data = pd.read_csv('https://raw.githubusercontent.com/pirandello/apriori/master/movie_dataset.csv', 
                         header = None)
```

* explore the dataset

```python
movie_data.T
```

```python
movie_data.shape
```

* transform dataframe to list of lists (suitable format for apyori)

```python
records = []
for i in range(1, len(movie_data[0])):
    records.append([str(movie_data.values[i,j]) for j in range(0, 20)])
```

* instantiate apriori and set the input params based on following constraints:
    * we want only movies that are purchased at least 40 times
    * the minimum confidence for the rules is 20%
    * the minumum lift is 3

```python
association_rules = apriori(records, min_support=0.0053, min_confidence=0.20, min_lift=3)
```

```python
association_rules = list(association_rules)
```

* how many association rules did we obtained ?

```python
len(association_rules)
```

* print the first association rule

```python
association_rules[0]
```

* convert association rules to DataFrame 
    * use these columns: title_1, title_2, support, confidence, lift

```python
list(association_rules[0][0])[0]
```

```python
# movies
association_rules[0][0]

# support value
association_rules[0][1]

# confidence value
association_rules[0][2][0][2]

# lift value
association_rules[0][2][0][3]
```

```python
results = pd.DataFrame()

for rule_idx in range(len(association_rules)):
    result = pd.DataFrame()
    
    result.loc[rule_idx, 'title_1'] = list(association_rules[rule_idx][0])[0]
    result.loc[rule_idx, 'title_2'] = list(association_rules[rule_idx][0])[1]
    result.loc[rule_idx, 'support'] = association_rules[rule_idx][1]
    result.loc[rule_idx, 'confidence'] = association_rules[rule_idx][2][0][2]
    result.loc[rule_idx, 'lift'] = association_rules[rule_idx][2][0][3]
    results = pd.concat([results, result], axis=0)
```
