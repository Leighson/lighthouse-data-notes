In NumPy, we have a 3-D array. The first 2 dimensions are 900x900. This represents the size of the picture. The last dimension is 3, therefore we have one layer for red, one for blue and one for green. Each value represents the intensity of a particular color in that specific position.

Representation of images in Numpy using NumPy arrays) allows us to do the simple changes to these images using mathematical operations.

```python
imgmat = np.array(myImage)

# storing the shape of picture into the variables
lx, ly, lz = imgmat.shape
```

```python
X,Y = np.ogrid[0:lx, 0:ly]
print(X.shape)
print(Y.shape)
```

```python
mask = (X -lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly /4
# this is the mathematical representation of a circle, with coordinates lx,ly and X,Y representing the boundaries
```

```python
# Where mask is True, insert black color
imgmat[:,:,0][mask] = 0
imgmat[:,:,1][mask] = 0
imgmat[:,:,2][mask] = 0
```

```python
plt.figure(figsize=(9,6))
plt.imshow(imgmat)
```
