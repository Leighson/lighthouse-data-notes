## Pictures

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
myImage = Image.open('data/flower_square.jpg')
plt.figure(figsize=(9, 6))
plt.imshow(myImage)
```

### Covert the image into a NumPy array

```python
imgmat = np.array(myImage)
plt.figure(figsize=(9, 6))
plt.imshow(imgmat)
```

```python
print(type(myImage))
print(type(imgmat))
print(imgmat.shape)
```

In NumPy, we have a 3-D array. The first 2 dimensions are 900x900. This represents the size of the picture. The last dimension is 3, therefore we have one layer for red, one for blue and one for green. Each value represents the intensity of a particular color in that specific position.

Representation of images in NumPy (using arrays) allows us to do simple changes to these images using mathematical operations.

### Unpacking the dimensions of an image

```python
imgmat = np.array(myImage)
lx, ly, lz = imgmat.shape
```

`ogrid` returns an open multi-dimensional *meshgrid*. Only one dimension of each returned array is greater than 1. The dimension and number of the output arrays are equal to the number of indexing dimensions.

```python
# storing the shape of a picture into the variables
X, Y = np.ogrid[0:lx, 0:ly]
print(X.shape)
print(Y.shape)
```

```python
# where mask is True, insert black color
mask = (X - lx / 2) ** 2 + (Y - ly / 2) ** 2 > lx * ly / 4

# go through each dimension and apply the mask
# the slice will specifically target each pixel that returns a True value
imgmat[:, :, 0][mask] = 0
imgmat[:, :, 1][mask] = 0
imgmat[:, :, 2][mask] = 0

plt.figure(figsize=(9, 6))
plt.imshow(imgmat)
```
