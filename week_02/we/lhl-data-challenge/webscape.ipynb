```python
with open('teams.json', 'r') as f:
    teams_json = json.load(f)
    
teams_df = pd.json_normalize(teams_json)

teams_df
```

```python
with open('points.json', 'r') as f:
    points_json = json.load(f)

points_df = pd.json_normalize(points_json)

points_df['prize_points'] = points_df['prize_points'].transform(int)
```

```python
teams = teams_df.merge(points_df, on='url')

teams_sorted = teams.sort_values(by = 'prize_points', ascending=False)

teams_sorted.loc[teams_sorted['private'] == False]
```
